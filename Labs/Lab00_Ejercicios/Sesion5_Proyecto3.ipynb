# Actividad Regresion Lineal
Crea un modelo de regresion lineal para predecir el salario basandose en la edad

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.DataFrame()
df["Edad"] = [25, 30, 35, 40, 45]
df["Salario"] = [3000, 3500, 4000, 4500, 5000]
df
```

```python
plt.scatter(df["Edad"], df["Salario"])
plt.grid()
plt.show()
```

```python
x = df[["Edad"]]
x
```

```python
y = df["Salario"]
y
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(x,y)
```

```python
modelo.predict([[47]])
```

```python
plt.plot(df["Edad"], modelo.predict(x))
plt.grid()
plt.show()
```

```python
plt.scatter(df["Edad"], df["Salario"], color='blue')
plt.plot(df["Edad"], modelo.predict(x), color='red')
plt.title('Relacion entre Edad y Salario')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid()
plt.show()
```
