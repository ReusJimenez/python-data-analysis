Seleccionen los colegios donde la propiedad de la planta física sea de una persona natural o comunidad religiosa.

```python
colegios_filtrados = df_colegios[ 
    (df_colegios['propiedad_Planta_Fisica'] == 'PERSONA NATURAL') | 
    (df_colegios['propiedad_Planta_Fisica'] == 'COMUNIDAD RELIGIOSA')
]
```

¿Hay alguno de estos colegios que sea para capacidades excepcionales?

```python
colegios_capacidades_excepcionales = colegios_filtrados[
    colegios_filtrados['capacidades_Excepcionales'].notnull()
]
colegios_capacidades_excepcionales
```

¿Cuáles son los colegios que ofrecen todos los niveles educativos? ¿Qué idiomas ofrecen?

```python
colegios_todos_niveles = df_colegios[
    df_colegios['niveles'] == 'PREESCOLAR,MEDIA,BÁSICA SECUNDARIA,BÁSICA PRIMARIA,PRIMERA INFANCIA'
]
colegios_todos_niveles[['nombreestablecimiento','niveles','idiomas']]
```
