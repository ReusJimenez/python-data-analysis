```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2.  Grafico de  lineas
#a. Crear el lienzo(Área de Gráfico)
plt.figure(figsize=(10,6)) #Tamaño del gráfico

#b.  Creación de listas  Año y población de país  "Perú"
x = [2018,2019,2020,2021,2022,2023] #Eje X (abajo)
y = [28,32,32,33,34,34] #Eje Y (datos dependientes)
#País de colombia
x2 =[2018,2019,2020,2021,2022,2023]
y2 =[41,43,44,44,48,49]

#c . establecer datos al gráfico plt
plt.plot(x,y, marker='o', linestyle='--', color='r', label='Perú')
plt.plot(x2,y2, marker='d', linestyle='-', color='b', label='Colombia')

#d. Personalizar  - Titulo principal, ejes X o y , leyenda
plt.legend(loc='lower right') #Agregando leyenda
plt.title('Población de perú y colombia') #Titulo principal
plt.xlabel('Años') #Titulos en los ejes  X   Y
plt.ylabel('Población')
plt.grid()
```

```python
#3. Graficos de columnas
# Comparacion de poblacion de varios países
x = ['Perú','Bolivia','Brasil','Chile'] #Países
y = [33,28,50,40] #Población en millones

#Agregar los datos al gráfico de columnas
plt.bar(x,y, label='Paises') #Gráfico de columnas

#Perzonalizar
plt.title('Población en países')
plt.xlabel('Países')
plt.ylabel('Población')
plt.legend(loc='upper left')
plt.show() #Mostrar gráfico
```

```python
#4. Gráfico Circular
colores =['r','g','b','y'] #lista de colores
plt.pie(y, labels=x, colors=colores)

#personalizar
plt.title('Población en países')
plt.legend(loc='center')
plt.show()
```

```python
#5. Histograma - Frecuencia
edades = [22,18,20,35,25,26,32,35,19,18]
bins = [15,20,25,30,35,40]

#crear el histograma
plt.hist(edades,bins, edgecolor='black')

#Personalizarlo
plt.title('Histograma de edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#6. Grafico de dispersión (importante para ML, regresión lineal)
tiempo = [1,2,3,4,5,6] #Eje X
nota = [8,9,14,16,18,20] #Eje Y
plt.scatter(tiempo, nota)
plt.grid() # Cuadriculas
plt.show()
```
