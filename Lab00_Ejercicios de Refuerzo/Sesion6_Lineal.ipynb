# **Sesión 6 - Regresión lineal**

```python
#1. Importar las librerias necesarias
import pandas as pd # para trabajar con el dataframe
import matplotlib.pyplot as plt # para graficar
from sklearn import linear_model # para la tecnica o el algoritmo de regresion lineal
```

```python
#2. Crear un dataset (dataframe) - Datos historicos
df = pd.DataFrame()
df["Area"] = [2600, 3000, 3200, 3600, 4000] # area en metros cuadrados - Terrenos industriales
df["Precio"] = [550000, 565000, 610000, 680000, 725000]  # precio de venta del terreno

#Datos etiquetados
df
```

```python
#3. Ver los datos en un grafico de dispersion
plt.scatter(df["Area"],df["Precio"])
plt.grid()
plt.show()
```

```python
#4. Crear un modelo (robot) de regresion lineal
#a. Obtener los datos del Area m2
x = df[["Area"]] # x es un dataframe, gracias a que se pasa una lista de columnas usando doble corchete
x
```

```python
type(x)
```

```python
#b. Obterner los precios en la variable y (Serie)
y = df["Precio"] # y es una serie, ASI DEBE SER, no un dataframe
y
```

```python
#c. Crear el modelo de regresion lineal
modelo = linear_model.LinearRegression()
#Ya tienes el modelo, ahora debes entrenarlo con los datos
```

```python
#5. Entrenar el modelo
modelo.fit(x,y)
```

```python
#6. Modelo realice la prediccion
modelo.predict([[3500]]) # 3500 m2
```

```python
#Conocer la mejore recta que el modelo a encontrado
plt.plot(df["Area"], modelo.predict(x))
plt.grid()
plt.show()
```

```python
#7. Graficar el modelo de regresion lineal
plt.scatter(df["Area"],df["Precio"])
plt.plot(df["Area"],modelo.predict(x),color="red")
plt.grid()
plt.show()
```
