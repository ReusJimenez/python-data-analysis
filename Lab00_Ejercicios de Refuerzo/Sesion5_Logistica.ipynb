# **Sesión 5 - Regresión logistica**

```python
#1. Importar las librerias
import pandas as pd #Leer, Explorar, Limpiar
import matplotlib.pyplot as plt #Gráficos estadísticos
#Libreria para el modelo o la técnica de regresión Logistca
from sklearn.linear_model import LogisticRegression
#Libreria para separar los datos de entrenamiento y prueba
from sklearn.model_selection import train_test_split
```

```python
#2. Leer los datos -->  en DataFrame
df = pd.read_csv('E:/datos_seguro.csv')
df
```

```python
#3. Exploración
#a. Identificando datos
df.head()
```

```python
#b. analisis previo (Gráfico)
plt.scatter(df.edad,df.compra)
plt.grid()
plt.show()
```

```python
#4. Separar en 2 conjuntos de datos
#90% entrenamiento   10% para probar
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(df[["edad"]], df["compra"], train_size=0.9)
```

```python
#4.1. Explorar los datos de entrenamiento
X_entrena
```

```python
#4.2. Explorar los datos de prueba  10%
X_prueba
```

```python
#5. Creación del Modelo "Regresión Logistica"
modelo = LogisticRegression()
```

```python
#6. Entrenar al Modelo
modelo.fit(X_entrena, y_entrena) #Brindar datos para que aprenda
```

```python
#7. Probar el modelo(Robot)
#7.1.  Registrar datos nuevos en un dataFrame
datos_nuevos = pd.DataFrame({"edad":[25,35,45,55]})
datos_nuevos
```

```python
#7.2. Probabilidad que compren el seguro
probabilidades = modelo.predict_proba(datos_nuevos)
probabilidades
```

```python
#7.3 Extraer las probabilidades que "si van a comprar"
prob_compra = probabilidades[:,1]
prob_compra
```

```python
#7.4.  Crear una visualización
plt.scatter(df.edad, df.compra)
plt.scatter(datos_nuevos["edad"], prob_compra, color="red")
```
